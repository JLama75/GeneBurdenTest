```python
df_vcf=pd.read_csv('../step1/output4_GT.tsv', sep='\t')
# df_vcf=df_vcf.drop(['#CHROM', 'POS', 'REF', 'ALT', 'INFO', 'FORMAT'], axis=1)
df_synonymous=pd.read_csv('../step2/list_synonymous.tsv')
df_synonymous=df_synonymous.drop_duplicates()
df_vcf_synonymous=df_vcf[df_vcf['ID'].isin(df_synonymous['variant'])]
df_synonymous_summary=pd.DataFrame()
df_synonymous_summary['IID']=df_vcf_synonymous.columns[1:]
list_count=[]
for column in df_vcf_synonymous.columns[1:]:
#     print(column)
    column_counts = df_vcf_synonymous[column].tolist().count('0/1')+df_vcf_synonymous[column].tolist().count('1/1')*2
    list_count.append(column_counts)
df_synonymous_summary['count_synonymous']=list_count
display(df_synonymous_summary)
```
